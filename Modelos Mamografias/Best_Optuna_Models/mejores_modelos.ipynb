```python
import torch
torch.cuda.memory_cached()
import os
import torch
import torch.nn as nn
import numpy as np
import torch.optim as optim
import pandas as pd
from torch.utils.data import Dataset, DataLoader, Subset
from torchvision import transforms
from PIL import Image
from sklearn.model_selection import train_test_split
import optuna
import sys
import matplotlib.pyplot as plt
import numpy as np
import cv2
import pandas as pd
from torchvision.transforms.functional import to_tensor, to_pil_image
import best_model_mamografia as bmg
import best_model_mamografia1 as bmg1
import best_model_mamografia2 as bmg2
import best_model_mamografia3 as bmg3
import best_model_mamografia4 as bmg4
import best_model_mamografia5 as bmg5
import best_model_mamografia6 as bmg6
import best_model_mamografia7 as bmg7
import best_model_mamografia8 as bmg8
import best_model_mamografia9 as bmg9
import best_model_mamografia10 as bmg10

sys.path.append(r"C:/Users/Matias/Desktop/MedTech//utils")
import creacion_df_torch as qol

dataset_dir = r"C:/Users/Matias/Desktop/MedTech/Dataset_Consolidado"

# Preparar datos
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
df = qol.crear_dataframe_binario(dataset_dir)
class_counts = df['etiqueta'].value_counts()
train_loader, val_loader, test_loader = qol.prepare_data_loaders(df, batch_size=100)

# Preparar los DataLoaders (usando el DataFrame binario como ejemplo)
# Guardar el DataFrame de prueba
_, test_val_df = train_test_split(df, test_size=0.3, random_state=46, stratify=df['etiqueta'])
_, test_df = train_test_split(test_val_df, test_size=0.33, random_state=46, stratify=test_val_df['etiqueta'])



# Seleccionar imágenes del conjunto de prueba
class_a_images = test_df[test_df['etiqueta'] == 'benign'].sample(n=5, random_state=46)
class_b_images = test_df[test_df['etiqueta'] == 'malignant'].sample(n=5, random_state=46)
selected_images = pd.concat([class_a_images, class_b_images])
```

# Modelos Binarios Cancer-No Cancer

```python
# Entrenar modelo
model_bmg = bmg.train_model(device, train_loader, val_loader)

# Evaluar modelo
test_accuracy = qol.evaluate_model(device, model_bmg, test_loader)
print(f'Final Test Accuracy: {test_accuracy:.4f}')

transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Crear una instancia de GradCAM
target_layer = model_bmg.features[-1]  # Última capa convolucional
grad_cam = qol.GradCAM(model_bmg, target_layer)


plt.figure(figsize=(20, 10))
for idx, (i, row) in enumerate(selected_images.iterrows()):
    ruta_imagen_prueba = row['ruta']
    true_class = row['etiqueta']

    # Cargar y preparar la imagen
    imagen = Image.open(ruta_imagen_prueba).convert('RGB')
    imagen_tensor = transform(imagen).unsqueeze(0).to(device)

    # Obtener predicción del modelo
    model_bmg.eval()
    with torch.no_grad():
        output = model_bmg(imagen_tensor)
        _, pred_class = torch.max(output, 1)
        pred_class = pred_class.item()

    # Generar el mapa de calor
    mask = grad_cam.generate(imagen_tensor)

    # Mostrar la imagen con el mapa de calor
    imagen_np = np.array(imagen.resize((256, 256))) / 255.0
    cam_image = qol.show_cam_on_image(imagen_np, mask)

    plt.subplot(5, 4, idx*2+1)
    plt.imshow(imagen_np)
    plt.title(f'Imagen Original\nClase: {true_class}')
    plt.axis('off')

    plt.subplot(5, 4, idx*2+2)
    plt.imshow(cam_image)
    plt.title(f'Grad-CAM\nPred: {pred_class}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
