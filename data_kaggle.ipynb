```python
import os
import shutil
import pandas as pd

# Cargar el dataframe (asegúrate de ajustar la ruta y el nombre del archivo)
df = pd.read_csv('C:/Users/Matias/Desktop/MedTech/RSNA Mammography PNG Dataset/train.csv')

# Definir las rutas de las carpetas
source_folder = "C:/Users/Matias/Desktop/MedTech/RSNA BCD 1024x512 Preprocessed/train_images"
destination_folder = "C:/Users/Matias/Desktop/MedTech/Dataset_Consolidado"

# Crear las carpetas de destino si no existen
benign_folder = os.path.join(destination_folder, 'benign')
malignant_folder = os.path.join(destination_folder, 'malignant')
os.makedirs(benign_folder, exist_ok=True)
os.makedirs(malignant_folder, exist_ok=True)

# Función para mover y renombrar las imágenes
def move_and_rename_images(row):
    image_id = row['image_id']
    patient_id = row['patient_id']
    cancer_status = 'benign' if row['cancer'] == 0 else 'malignant'
    laterality = row['laterality']
    view = row['view']
    
    # Definir el nombre nuevo de la imagen
    new_image_name = f"{patient_id}_{laterality}_{view}_{cancer_status}_{image_id}.png"
    
    # Definir las rutas de origen y destino
    source_image_path = os.path.join(source_folder, str(patient_id), f"{image_id}.png")
    if cancer_status == 'benign':
        dest_image_path = os.path.join(benign_folder, new_image_name)
    else:
        dest_image_path = os.path.join(malignant_folder, new_image_name)
    
    # Mover y renombrar la imagen
    if os.path.exists(source_image_path):
        shutil.move(source_image_path, dest_image_path)
    else:
        print(f"Imagen no encontrada: {source_image_path}")

# Aplicar la función a cada fila del dataframe
df.apply(move_and_rename_images, axis=1)
```
